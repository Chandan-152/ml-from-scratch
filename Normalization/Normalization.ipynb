#import minmaxscaler libraries

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler=MinMaxScaler()
```

```python
height_scaled=scaler.fit_transform(df[['Height']])
weight_scaled=scaler.fit_transform(df[['Weight']])
```

#data put into dataframe

```python
height_scaled=pd.DataFrame(height_scaled,columns=['height_scaled'])
weight_scaled=pd.DataFrame(weight_scaled,columns=['weight_scaled'])
```

```python
df=pd.concat([height_scaled,weight_scaled],axis=1)
```
