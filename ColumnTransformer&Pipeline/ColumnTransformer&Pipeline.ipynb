## ðŸ“¦ Import Required Libraries



```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv("/content/House_Rent_Dataset.csv")
```

```python
df.sample(5)
```

```python
df=df.drop(columns=['Posted On','Floor','Area Locality'])
```

```python
df.sample(5)
```

```python
x=df.drop(columns=['Rent'])
```

```python
y=df['Rent']
```

## ðŸ”€ Trainâ€“Test Split



```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
x_train.sample(10)
```

## ðŸ§© ColumnTransformer: Categorical Feature Encoding



```python
from sklearn.preprocessing import OneHotEncoder,OrdinalEncoder
```

```python
from sklearn.compose import ColumnTransformer
transformer=ColumnTransformer(transformers=(
    ('trfn1',OrdinalEncoder(categories=[['Built Area','Carpet Area','Super Area'],
                                       ['Unfurnished','Semi-Furnished','Furnished']]),
                                       ['Area Type','Furnishing Status']),
    ('trfn2',OneHotEncoder(sparse_output=False,drop='first'),['City','Tenant Preferred','Point of Contact'])
),remainder='passthrough')
```

```python
transformer
```

```python
x_train_transformer=transformer.fit_transform(x_train)
```

```python
x_test_transformer=transformer.transform(x_test)
```

## ðŸ”— Preprocessing Pipeline & Model Serialization


```python
from sklearn.pipeline import Pipeline,make_pipeline
```

```python
pipeline=Pipeline([
    ('trfn',transformer)
])
```

```python
pipeline.fit(x_train,y_train)
```

```python
pipeline
```

```python
pipeline.named_steps['trfn'].transformers_
```
